```python
df['Группа компонента'] = df.apply(lambda row: 'Сборка' if row['Наименование компоненты'] in ["Кузов","Каркас", "Салон", "Подвеска", "Двигатель", "Трансмиссия"] else 'Компонент', axis=1)
df
```

```python
krepezh = {'Болт', 'Гайка', 'Штифт', 'Шайба', 'Шуруп'}
decor = {'Коврики', 'Подушки'}
```

```python
def categorize(row):
    if row['Наименование компоненты'] in krepezh:
        return 'Крепеж'
    elif row['Наименование компоненты'] in decor:
        return 'Декор'
    elif 'Сборка' in row['Группа компонента']:
        return 'Сборка'
    else:
        return 'Детали'
```

```python
df['Категория'] = df.apply(categorize, axis=1)
df
```

```python
df['Код компоненты'].str.findall(r'[A-Za-z]')
```

```python
def proizvoditel(component):
    if component.isnumeric() or re.match(r'[A-Za-z]', component):
        return 'Иностранное'
    elif 'ГОСТ' in component:
        return 'РФ по ГОСТ'
    elif any(char.isalpha() for char in component):
        return 'РФ'

df['Производитель'] = df['Код компоненты'].apply(proizvoditel)
df
```

# Сравнение файлов Машина 1 и Машина 2

```python
changes_delete = []
changes_add = []
changes_rename = []
```

```python
for index, row in rf.iterrows():
    if row.to_list() not in cf.values.tolist():
        changes_delete.append(f"Элемент {row.to_list()} удален")
```

```python
for index, row in cf.iterrows():
    if row.to_list() not in rf.values.tolist():
        changes_add.append(f"Элемент {row.to_list()} добавлен")
```

```python
for column in rf.columns:
    if rf[column].count() != cf[column].count():
        changes_rename.append(f"Изменилось количество элементов в столбце {column}")
```

```python
for change in changes_delete:
    print(change)
print('\n')
for change in changes_add:
    print(change)
print('\n')
for change in changes_rename:
    print(change)
print('\n')
```

```python
m = (rf.merge(cf, how='outer', on=['Код изделия','Наименование изделия','Код сборки','Наименование сборки','Код компоненты','Наименование компоненты','Кол-во\nна\nсборку'], 
              suffixes=['', '_new'], indicator=True))
m2 = (cf.merge(rf, how='outer', on=['Код изделия','Наименование изделия','Код сборки','Наименование сборки','Код компоненты','Наименование компоненты','Кол-во\nна\nсборку'], 
              suffixes=['', '_new'], indicator=True))
```

```python
m3=pd.merge(m.query("_merge=='right_only'"), m2.query("_merge=='right_only'"), how ='outer').drop_duplicates(subset=['Код изделия','Наименование изделия','Код сборки','Наименование сборки','Код компоненты','Наименование компоненты','Кол-во\nна\nсборку'])
m3
```
